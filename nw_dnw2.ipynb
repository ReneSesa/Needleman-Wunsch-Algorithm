```python

NUMBER_OF_ROWS = len(sequence1) + 1 
NUMBER_OF_COLUMNS = len(sequence2) + 1
row_labels = [label for label in "-"+sequence1]
column_labels = [label for label in "-"+sequence2]


scoring_array=np.zeros((NUMBER_OF_ROWS,NUMBER_OF_COLUMNS), dtype=float)



traceback_array = full([NUMBER_OF_ROWS,NUMBER_OF_COLUMNS],"-")


arrow = "-"

match_bonus = 1
gap_penalty=0.1
```

```python
for row in range(NUMBER_OF_ROWS):
    for col in range(NUMBER_OF_COLUMNS):    
        #set first row and column to zero
        if row ==0 and col == 0:
            score=0
            arrow='-'
        elif row == 0:

            previous_score = scoring_array[row,col - 1]
            
            score = previous_score + gap_penalty
            arrow = left_arrow
        elif col == 0:
            
            previous_score = scoring_array[row -1,col]
            score = previous_score + gap_penalty
            arrow = up_arrow
            
        else: 

            index1=ord(sequence1[row-1])-65
            index2=ord(sequence2[col-1])-65

            
            if(index1>=index2):
                alphabet_score=float(File_data[index1][index2])
                #print("alphabet_score "+ str(alphabet_score))
            else:
                alphabet_score=float(File_data[index2][index1])
                #print("alphabet_score "+ str(alphabet_score))
            
            cost=min(alphabet_score, gap_penalty)
            ################################
            
            cell_to_the_left = scoring_array[row,col-1]
            from_left_score = cell_to_the_left + cost#alphabet_score
            #print("from_left_score "+str(from_left_score))
             
            
            above_cell = scoring_array[row-1,col]
            from_above_score = above_cell +cost# alphabet_score
            #print("from_above_score "+str(from_above_score))
            
            diagonal_left_cell = scoring_array[row-1,col-1]
            

            if sequence1[row-1] == sequence2[col-1]:
                diagonal_left_cell_score = diagonal_left_cell + match_bonus
            else:
                diagonal_left_cell_score = diagonal_left_cell
                
            #print("diagonal_left_cell_score "+str(diagonal_left_cell_score))
            
            score = max([from_left_score,from_above_score,diagonal_left_cell_score])       
            #print("score is: " + str(score))

            
            #traceback path of current element
            if score == diagonal_left_cell_score:
                arrow = up_left_arrow            
            elif score == from_left_score:
                arrow = left_arrow     
            elif score == from_above_score:
                arrow = up_arrow                

                
                
        #print("input score "+str(score))
        traceback_array[row,col]=arrow    
        scoring_array[row,col] = score
        #print("saved score " + str(scoring_array[row,col]))

#traceback_array[0,1]='-'
#traceback_array[1,0]='-' 
display(table_from_array(scoring_array,row_labels,column_labels))  
display(table_from_array(traceback_array,row_labels,column_labels))            
```

```python
def traceback_alignment(traceback_array,seq1,seq2,up_arrow = "\u2191" ,\
                        left_arrow="\u2190",up_left_arrow="\u2196",stop="-"):


    NUMBER_OF_ROWS = len(sequence1) + 1 
    NUMBER_OF_COLUMNS = len(sequence2) + 1 
    
    row = len(sequence1)
    col = len(sequence2)
    arrow = traceback_array[row,col]
    aligned_seq1 = ""
    aligned_seq2 = ""
    alignment_indicator = ""
    while arrow != "-":

            arrow = traceback_array[row,col]

            
            if arrow == up_arrow: 


                aligned_seq2 = "-"+aligned_seq2 
                aligned_seq1 = sequence1[row-1] + aligned_seq1
                alignment_indicator = " "+alignment_indicator
                row -=1
                            
            elif arrow == up_left_arrow:


                seq1_character = sequence1[row-1]
                seq2_character = sequence2[col-1]
                aligned_seq1 = sequence1[row-1] + aligned_seq1
                aligned_seq2 = sequence2[col-1] + aligned_seq2
                if seq1_character == seq2_character:
                    alignment_indicator = "|"+alignment_indicator
                else:
                    alignment_indicator = " "+alignment_indicator
                row -=1
                col -=1
                
            elif arrow == left_arrow:

                aligned_seq1 = "-"+aligned_seq1
                aligned_seq2 = sequence2[col-1] + aligned_seq2
                alignment_indicator = " "+alignment_indicator
                col -=1
                
            elif arrow == stop:
                break
                
                
    print(aligned_seq1)
    print(alignment_indicator)
    print(aligned_seq2)

            
    return aligned_seq1,aligned_seq2


traceback_alignment(traceback_array,sequence1,sequence2)

print(scoring_array[NUMBER_OF_ROWS-1][NUMBER_OF_COLUMNS-1])
```

```python
#init
n=len(sequence1)
m=len(sequence2)
scoring_arr=np.zeros((n+1,m+1), dtype=float)  #scoring array

for i in range(1,n+2):
    scoring_arr[i-1,0]=gap_penalty*(i-1)
    
for i in range(1,m+2):
    scoring_arr[0,i-1]=gap_penalty*(i-1)    
    
    
gradient_arr=np.zeros((n+2,m+2), dtype=float)  #gradient array 
gradient_arr[n+1,m+1]=1.

backtracking_arr=np.zeros((n+2,m+2,3), dtype=float)  #backtracking arrray
backtracking_arr[n+1,m+1,1]=1.

print(scoring_arr)
print("\n")
print(gradient_arr)
print("\n")
print(backtracking_arr)
```

```python
#filing substition matrix by comparing seq1 and seq2
substitution_matrix=np.zeros((n,m), dtype=float)  
for i in range(n):
    for j in range(m):
        if sequence1[i]!=sequence2[j]:
            index1=ord(sequence1[i-1])-65
            #print(index1)
            index2=ord(sequence2[j-1])-65
            #print(index2)
            
            if(index1>=index2):
                alphabet_score=float(File_data[index1][index2])
                #print("alphabet_score "+ str(alphabet_score))
            else:
                alphabet_score=float(File_data[index2][index1])
                #print("alphabet_score "+ str(alphabet_score))            
            
            
            substitution_matrix[i,j]=alphabet_score  
        else:                           
            substitution_matrix[i,j]=1
    

row_labels = [label for label in sequence1]
column_labels = [label for label in sequence2]

display(table_from_array(substitution_matrix,row_labels,column_labels))
```

```python
#smooth maximum
import math
def logsumexp(x, gamma):
    sum=0
    for i in x:
        sum+=math.exp(i*gamma)
    return (1/gamma)*(math.log(sum))
```

```python
#soft maximum
#mo argument still not implemented, ignore it

def max_argmax(x, gamma, mo):
    m=logsumexp(x,gamma)
    q=[]
    for element in x:
        q.append(math.exp((element/gamma)-(m/gamma)))
    return q
```
